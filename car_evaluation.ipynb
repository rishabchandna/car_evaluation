```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('car.data',header=None)
df.head()
```

```python
df.rename(columns={0:'buying_price',1:'maintenance_cost',2:'no_of_doors',3:'no_of_persons',4:'lug_boot',5:'safety',6:'decision'},inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
buying_price = pd.get_dummies(df['buying_price'],drop_first=True,prefix='buying_price')
```

```python
maintenance_cost = pd.get_dummies(df['maintenance_cost'],drop_first=True,prefix='maintenance_cost')
maintenance_cost
```

```python
df = pd.concat([df,maintenance_cost],axis=1)
df.head()
```

```python
lug_boot = pd.get_dummies(df['lug_boot'],drop_first=True,prefix='lug_boot')
lug_boot
```

```python
df = pd.concat([df,lug_boot],axis=1)
df.head()
```

```python
safety = pd.get_dummies(df['safety'],drop_first=True,prefix='safety')
safety.head()
```

```python
df = pd.concat([df,safety],axis=1)
df.head()
```

```python
df = pd.concat([df,buying_price],axis=1)
```

```python
df.head()
```

```python
df.drop(columns=['buying_price','maintenance_cost',"lug_boot",'safety'],inplace=True)
```

```python
df.decision.unique()
```

```python
map_decision = {'unacc':1, 'acc':2, 'vgood':3, 'good':4}
```

```python
df['decision'] = df.decision.map(map_decision)
df.head()
```

```python
doors = pd.get_dummies(df['no_of_doors'],drop_first=True,prefix='no_of_doors')
doors.head()
```

```python
df = pd.concat([df,doors],axis=1)
```

```python
persons = pd.get_dummies(df['no_of_persons'],drop_first=True,prefix='no_of_persons')
persons.head()
```

```python
df = pd.concat([df,persons],axis=1)
df.head()
```

```python
df.drop(columns=['no_of_doors','no_of_persons'],inplace=True)
```

```python
df.head()
```

## Train-Test Split

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['decision'])
y = df.pop('decision')
```

```python
# standardizing all the variables using standard scaler
```

```python
y.head()
```

```python
X.head()
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
y_train.shape
```

### Model Evaluation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsOneClassifier,OneVsRestClassifier
```

```python
LR = LogisticRegression()
onevsrest = OneVsRestClassifier(LR)
onevsrest.fit(X_train_scaled,y_train)
#onevsrest.summary()
```

```python
###making prediction on test set
prediction_onevsRest =onevsrest.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_test,prediction_onevsRest)
```

```python
print(classification_report(y_test,prediction_onevsRest))
```
